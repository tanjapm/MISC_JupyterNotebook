```python
time = np.linspace(0,2000,50)

x0  = np.linspace( 1e-2, 2,50 )

Temp = np.array( [ np.log( x ) for x in x0 ] )

#plt.figure();
fig, ax = plt.subplots(figsize = (12,6))
plt.plot(time,Temp,'k',linewidth = 1.5)


# plot 
xdots = [270,850,920,1490]
ydots = [-1.25,-0.15,-0.05,0.4]
ydots_scale = [-0.5,-5.5,-16,2.2]

for i in range(4):
  #horizontal lines
    plt.axhline(ydots[i],xmax= float(xdots[i]) /2000,linestyle='--', color='grey')
  #vertical lines
    plt.axvline(xdots[i],ymax= float(ydots[i])*ydots_scale[i],linestyle='--', color='grey')
    plt.plot(xdots[i],ydots[i],'o',color='g')


xlabels = np.arange(0,2500,500)
ylabels = np.arange(0 , 0.7, 0.1)
plt.locator_params(axis='x', nticks=5)
plt.locator_params(axis='y', nticks=8)
ax.set_xticklabels(xlabels)
ax.set_yticklabels(ylabels)
plt.ylim(-5,1)
plt.xlim(0,2500)
plt.xlabel('Time [ms]',fontsize = 14);
plt.ylabel('Dimensionless temperature [-]',fontsize = 14);

plt.savefig('ignitionprocess.png')
```
